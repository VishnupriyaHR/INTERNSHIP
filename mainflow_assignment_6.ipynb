```python
df=df.drop_duplicates()
df.shape
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df[['title', 'release_year', 'rating']].head()
```

```python
# Explore the distribution of release years:
df['release_year'].value_counts().plot(kind='bar')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Distribution of Release Years')
plt.show()
```

```python
# Explore the distribution of ratings:
df['rating'].value_counts().plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Explore the relationship between release year and rating:
sns.scatterplot(data=df, x='release_year', y='rating')
plt.xlabel('Release Year')
plt.ylabel('Rating')
plt.title('Release Year vs. Rating')
plt.show()
```

```python
# Boxplot of release year by rating
sns.boxplot(x='rating', y='release_year', data=df)
plt.xlabel('Rating')
plt.ylabel('Release Year')
plt.title('Release Year by Rating (Boxplot)')
plt.show()
```

```python

# Violin plot of release year by rating
sns.violinplot(x='rating', y='release_year', data=df)
plt.xlabel('Rating')
plt.ylabel('Release Year')
plt.title('Release Year by Rating (Violin Plot)')
plt.show()
```

```python
# Create a histogram of release years
sns.histplot(data=df, x='release_year', bins=10)
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.title('Distribution of Release Years (Histogram)')
plt.show()
```

```python
# Pie chart of ratings
labels = df['rating'].value_counts().index.to_list()
sizes = df['rating'].value_counts().to_list()
colors = sns.color_palette('husl', len(labels))

plt.pie(sizes, labels=labels, colors=colors, autopct='%.1f%%')
plt.title('Distribution of Ratings (Pie Chart)')
plt.show()
```

```python
sns.scatterplot(data=df, x='release_year', y='rating')
plt.xlabel('Release Year')
plt.ylabel('Rating')
plt.title('Release Year vs. Rating')
plt.show()
```

```python

df.groupby('release_year')['title'].count().plot(kind='line')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.title('Number of Titles Released per Year')
plt.show()
```
